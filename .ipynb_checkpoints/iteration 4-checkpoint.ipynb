```python
import findspark
findspark.init('/home/ubuntu/spark-2.1.1-bin-hadoop2.7')
import pyspark
from pyspark.sql import *
spark = SparkSession.builder.appName('dataframe_transformation').getOrCreate()
spark = SparkSession.builder.appName('pandasToSparkDF').getOrCreate()
```

```python
from pyspark.sql.types import *

mySchema = StructType([ StructField("Suburb", IntegerType(), True)\
                       ,StructField("Rooms", IntegerType(), True)\
                       ,StructField("Type", IntegerType(), True)\
                       ,StructField("Price", FloatType(), True)\
                       ,StructField("Distance", FloatType(), True)\
                       ,StructField("Bathroom", FloatType(), True)\
                       ,StructField("Car", FloatType(), True)\
                       ,StructField("Landsize", FloatType(), True)\
                       ,StructField("BuildingArea", FloatType(), True)\
                       ,StructField("YearBuilt", FloatType(), True)\
                       ,StructField("CouncilArea", IntegerType(), True)\
                       ,StructField("Lattitude", FloatType(), True)\
                       ,StructField("Longtitude", FloatType(), True)\
                       ,StructField("Regionname", IntegerType(), True)\
                       ,StructField("Propertycount", FloatType(), True)\
                       ,StructField("Age", FloatType(), True)])
```

```python
df = spark.createDataFrame(data,schema=mySchema)
```

```python
# Import VectorAssembler and Vectors
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler
```

```python
# The input columns are the feature column names, and the output column is what you'd like the new column to be named. 
assembler = VectorAssembler(
    inputCols=["Suburb","Rooms", 
               "Type",'Distance',
              "Bathroom","Car",
              "Landsize","BuildingArea",
              'YearBuilt',"CouncilArea",
              "Lattitude","Longtitude",
              "Regionname","Propertycount",
              "Age"],
    outputCol="features")
```

```python
# Now that we've created the assembler variable, let's actually transform the data.
output = assembler.transform(df)
```

```python
final_data = output.select("features",'Price')
final_data.printSchema()
```

```python
train_data,test_data = final_data.randomSplit([0.8,0.2])
```

```python
# Let's see our training data.
train_data.describe().show()

# And our testing data.
test_data.describe().show()
```

```python
import findspark
findspark.init('/home/ubuntu/spark-2.1.1-bin-hadoop2.7')
spark = SparkSession.builder.appName('linear_regression_adv').getOrCreate()

from pyspark.ml.regression import LinearRegression
```

```python
lr = LinearRegression(labelCol='Price')
```

```python
# Fit the model to the data.
lrModel = lr.fit(train_data)
```

```python
# Let's evaluate the model against the test data.
test_results = lrModel.evaluate(test_data)
```

```python
# Interesting results! This shows the difference between the predicted value and the test data.
test_results.residuals.show()
```

```python
# Let's get some evaluation metrics (as discussed in the previous linear regression notebook).
print("RSME: {}".format(test_results.rootMeanSquaredError))
```

```python
# Print the coefficients and intercept for linear regression.
print("Coefficients: {} Intercept: {}".format(lrModel.coefficients,lrModel.intercept))
```

```python
# We can also get the R2 value. 
print("R2: {}".format(test_results.r2))
```

```python
spark = SparkSession.builder.appName('tree_methods_adv').getOrCreate()
```

```python
from pyspark.ml.regression import DecisionTreeRegressor,RandomForestRegressor
from pyspark.ml import Pipeline
from pyspark.ml.evaluation import RegressionEvaluator
```

```python
dtc = DecisionTreeRegressor(labelCol='Price', featuresCol='features', maxDepth = 20)
```

```python
# Train the models (it's three models, so it might take some time).
dtc_model = dtc.fit(train_data)
```

```python
dtc_predictions = dtc_model.transform(test_data)
```

```python
# Let's do something a bit more complex in terms of printing, just so it's formatted nicer. 
evaluator = RegressionEvaluator(
    labelCol="Price", predictionCol="prediction", metricName="rmse")
```

```python
rmse_dtc = evaluator.evaluate(dtc_predictions)
print("Root Mean Squared Error (RMSE) on test data = %g" % rmse_dtc)
```

```python
rfc = RandomForestRegressor(labelCol='Price',featuresCol='features',maxDepth=6, numTrees=30)
```

```python
rfc_model = rfc.fit(train_data)
```

```python
rfc_predictions = rfc_model.transform(test_data)
```

```python
rmse_rfc = evaluator.evaluate(rfc_predictions)
print("Root Mean Squared Error (RMSE) on test data = %g" % rmse_rfc)
```
